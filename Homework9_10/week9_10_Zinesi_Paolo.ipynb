The python script "hopfield.py" has been slightly modified in order to be run from the command line with extra controls.
Images of different simulations are shown and commented in this notebook.

A typical run of the script is performed as:

```command-line 
hopfield.py <INPUT_IMAGE> <block_animation> <verbose>
```

As an example, to run the script with the starting image "superman.jpg" without showing the animation, the following command is needed:
```command-line 
hopfield.py inputs/superman.jpg True True
```

The `verbose` specification allows to hide prints performed during the training part.

## Input = "batman0.png"
In this case the network has troubles in finding the correct matching with the superman logo because the shape of the logo is slightly different w.r.t. the one present during training. For this reason, the state of the network evolves to a final state that is intermediate between the cat image and the batman logo (spurious local minimum). Probably, the shape of the cat's ears is a good match of the top part of the image, good enough to confuse the network. The same final pattern is reached for similar reasons also for the images:

- **"batman01.png"** (the image is the same of "batman0.png" but with opposite colors)
- **"batman5.png"** (only the outer circle is present, and it is difficult to retrieve the whole image)

````python
%run hopfield.py inputs/batman0.png True True
````

## Input = "batman2.png"
For this image the network finds the correct pattern, because the input image is only a blurried version of the pattern stored previously. Since the overlap is significantly different from zero even if one image is blurried, the input image is already very close to the batman logo stored as a pattern and it evolves quickly to the desired pattern. The same final pattern is reached also for the images:
- **"batman3.png"** (the whole logo is present, except the outer circle)
- **"batman4.png"** (same as "batman3.png" but with inverted colors)

````python
%run hopfield.py inputs/batman2.png True False
````

## Input = "cat2.jpg"
The network is very good at finding the correct pattern when a portion of the original image is given as a starting point. In fact, any update that changes the portion of the original image is rejected and the starting image evolves to match the stored pattern. The same thing happens also for the image "cat3.jpg", which is another portion of the original cat pattern.

````python
%run hopfield.py inputs/cat2.jpg True False
````

## Input = "superman.jpg"
For this image, there is no learnt pattern that is able to match it. Thus, the input evolves to the same spurious minimum of "batman0.png".

````python
%run hopfield.py inputs/superman.jpg True False
````

## Input = "samir.jpg"
The network correctly learns the patter of human with raised hand, even if the two images are intrinsically different (one image is not obtained by transforming the other). We can this affirm that a true learning process is happening in the background, that is probably triggered by the recognition of a darker region in the bottom of the image (the shoulders of the human) and a smaller dark region on the bottom left of the image (the raised hand). As it can be seen in the next image ("paolo.jpg"), however, the presence of the hand is not mandatory in the recognition of the pattern.

````python
%run hopfield.py inputs/samir.jpg True False
````
